```python
import matplotlib.pyplot as plt

вартість = [2422189, 1121866, 425000, 997230, 764280, 385500, 483700, 375300, 363200, 583430, 937600, 829900]
споживання_палива = [80, 70, 85, 73, 92, 84, 79, 78, 89, 95, 94, 88]

plt.plot(вартість, споживання_палива, marker='o')

plt.xlabel('Вартість, грн.')
plt.ylabel('Споживання палива, л')
plt.title('Залежність споживання палива від вартості')
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

speed = [350, 180, 190, 210, 300, 285, 180, 250, 290, 303, 295, 280]
fuel_consumption = [80, 70, 85, 73, 92, 84, 79, 78, 89, 95, 94, 88]

plt.scatter(speed, fuel_consumption)

plt.xlabel('Максимальна швидкість, км/год')
plt.ylabel('Споживання палива, л')
plt.title('Залежність споживання палива від швидкості')

plt.show()
```

```python
import matplotlib.pyplot as plt

вантажопідйомність = [2910, 2350, 1800, 1950, 2150, 2835, 1950, 1970, 1834, 1873, 1745, 2400]
швидкість = [350, 180, 190, 210, 300, 285, 180, 250, 290, 303, 295, 280]

plt.scatter(швидкість, вантажопідйомність)
plt.xlabel('Максимальна швидкість, км/год')
plt.ylabel('Вантажопідйомність, кг')
plt.title('Залежність вантажопідйомності від швидкості')
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

speed = [350, 180, 190, 210, 300, 285, 180, 250, 290, 303, 295, 280]

plt.hist(speed, bins=5, edgecolor='black')

plt.xlabel('Швидкість, км/год')
plt.ylabel('Частота')
plt.title('Розподіл автомобілів за швидкістю')

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Марка 1', 'Марка 2', 'Марка 3', 'Марка 4', 'Марка 5', 'Марка 6', 'Марка 7', 'Марка 8', 'Марка 9', 'Марка 10', 'Марка 11', 'Марка 12']
speeds = [350, 180, 190, 210, 300, 285, 180, 250, 290, 303, 295, 280]

plt.boxplot(speeds)

plt.xlabel('Марки автомобілів')
plt.ylabel('Швидкість, км/год')

plt.title('Розподіл швидкостей різних марок автомобілів')

plt.xticks(range(1, len(labels) + 1), labels, rotation=45)

plt.show()
```
