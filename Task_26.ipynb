```python
from math import comb
```

```python
def probability_within_range(p, n, k_min, k_max):
    probability = 0.0
    for k in range(k_min, k_max + 1):
        probability += comb(n, k) * p**k * (1 - p)**(n - k)
    return probability
```

```python
p = 0.9  
n = 100  
k_min = 85  
k_max = 95
```

```python
result = probability_within_range(p, n, k_min, k_max)
```

```python
print("Ймовірність: {:.4f}".format(result))
```

```python
import matplotlib.pyplot as plt
```

```python
import statistics
```

```python
data = [10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12]
```

```python
frequency = {}
for value in data:
    if value in frequency:
        frequency[value] += 1
    else:
        frequency[value] = 1
```

```python
sorted_values = sorted(frequency.keys())
```

```python
x = sorted_values
y = [frequency[value] for value in sorted_values]
plt.plot(x, y, marker='o')
plt.xlabel('Значення')
plt.ylabel('Частота')
plt.title('Полігон частот F*(x)')
plt.show()
```

```python
mode = max(frequency, key=frequency.get)
```

```python
median = statistics.median(data)
```

```python
print("Мода: ", mode)
```

```python
print("Медіана: ", median)
```
